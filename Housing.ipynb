```python
# 一个完整的机器学习项目

import pandas as pd
```

```python
def load_housing_data():
    csv_path = "./datasets/housing/housing.csv"
    return pd.read_csv(csv_path)
```

```python
housing = load_housing_data()
```

```python
# 查看该数据集的前五行
housing.head()
```

```python
# 快速查看数据的描述
housing.info()

# 运行后可以发现total_bedrooms那一列只有20433，说明有207个街区缺少这个值
```

```python
# 查看某一个具体的项中都有哪些类别，每一个类别包含多少数目
housing["ocean_proximity"].value_counts()
```

```python
# 数值属性的概述，空值被忽略了
housing.describe()
```

```python
# 另一种了解数据类型的方法是画出每个数值属性的柱状图
import matplotlib.pyplot as plt
housing.hist(bins=50, figsize=(20, 20)) # 本来是(20, 50)的，但是在我电脑上显示效果不行
plt.show()
```
